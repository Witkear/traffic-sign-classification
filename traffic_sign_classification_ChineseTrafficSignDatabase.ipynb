# Global imports

```
import cv2
import os
import numpy as np

from tensorflow                    import one_hot
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers       import Flatten, Dense, Conv2D, Dropout
from tensorflow.keras.models       import Model, load_model
from tensorflow.keras.callbacks    import ModelCheckpoint
```

# Create dataset.

Inputs:

1. Path to train directory.

2. Path to test directory.

3. Number of classes.

Outputs:

1. train_images[numpy array]

2. train_labels[numpy array]

3. test_images[numpy array]

4. test_labels[numpy array]

```
def get_label( filename ):
  """Read label from filename."""
  cls_id = int( filename.split( "_" )[ 0 ] )
  return cls_id

def get_array( path ):
  """Reads each images from a directory and store as a numpy array."""
  images_list = [ ]
  labels_list = [ ]
  flag        = True
  for filename in os.listdir( path ):
    image = cv2.imread( os.path.join( path, filename ) )
    try:
      image = cv2.resize( image, ( 96, 96 ) )
      #Get label from filename.
      label = get_label( filename )
      #Each image is added as last element of the input list.
      images_list.append( image )
      labels_list.append( label )
    except:
      pass
  return np.array( images_list ), np.array( labels_list )

def create_dataset( train_path  = "data/train",
                    test_path   = "data/test",
                    num_classes = 58 ):
  """A function that creates dataset in suitable format"""

  train_X, train_y = get_array( train_path )
  test_X, test_y   = get_array( test_path )
  return train_X, train_y, test_X, test_y

train_X, train_y, test_X, test_y = create_dataset( )
```

# Observe dataset.

```
print( train_X.shape )
print( train_y.shape )
print( train_y[ 0 ] )
print( test_X.shape )
print( test_y.shape )
```

# Create model.

Load mobilenet as basemodel.

```
basemodel = MobileNetV2( input_shape = ( 96, 96, 3 ), 
                        include_top = False,
                        weights = "imagenet" )
```

# Check the summary.

```
basemodel.summary( )
```

# Freeze layer.

```
for layer in basemodel.layers[ :100 ]:
  layer.trainable = False
```

# Define model checkpoint.

```
!mkdir models
```

```
callback = ModelCheckpoint( 'models/weights.{epoch:02d}-{val_loss:.2f}.hdf5', monitor= "val_loss" )
```

# Create model.

```
!rm /content/workspace/models/*

in_to_custom = basemodel.output
x            = Flatten( name = "flatten_to_dense" )( in_to_custom )
x            = Dropout( 0.2 )( x )
x            = Dense  ( 128, name = "dense_hidden_0" )( x )
x            = Dropout( 0.2 )( x )
x            = Dense( 58, activation = "softmax", name = "dense_final" )( x )
model        = Model( inputs = basemodel.input, outputs = x )
print( model.summary( ) )
model        . compile( loss = "sparse_categorical_crossentropy", 
                        optimizer = "adam",
                        metrics = "accuracy" )
history      = model.fit( train_X, train_y, epochs = 100, 
                          callbacks = [ callback], validation_split = 0.2 )
```
